#### `ðŸ“Œ Feature Transformation: Deriving features/columns`

#### `BMI from height and weight`

```python
ol['bmi'] = ol['weight'] / (ol['height'] / 100) ** 2
```

```python
ol.head(8)
```

```python
ol['ratio'] = ol['height'] / ol['weight']
```

```python
ol.head(8)
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'

ol['bmi_category'] = ol['bmi'].apply(bmi_category)
```

```python
ol.head(8)
```

####  `Derived features/columns: bmi, bmi_category and ratio`

```python
ol['sex'] = ol['sex'].str.capitalize()
ol['team'] = ol['team'].str.capitalize()
ol['season'] = ol['season'].str.capitalize()
ol['city'] = ol['city'].str.capitalize()
ol['sport'] = ol['sport'].str.capitalize()
ol['medal'] = ol['medal'].str.capitalize()
ol['bmi_category'] = ol['bmi_category'].str.capitalize()
```

```python
ol.head(4)
```

```python
ol['age'] = pd.to_numeric(ol['age'], errors='coerce')
ol['height'] = pd.to_numeric(ol['height'], errors='coerce')
ol['weight'] = pd.to_numeric(ol['weight'], errors='coerce')
ol['year'] = pd.to_numeric(ol['year'], errors='coerce')
```
